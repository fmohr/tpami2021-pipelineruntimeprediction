```python
def fillUpCSVLearnResultFiles(file1, file2, outfile):
    df1 = pd.read_csv(file1)
    df2 = pd.read_csv(file2)
    j = df1.merge(df2, on=["openlid", "trainpoints", "seed", "classifier"])
    print(len(j))
```

```python
def mergeCSVLearnResultFiles(file1, file2, outfile):
    df1 = pd.read_csv(file1)
    df2 = pd.read_csv(file2)
    v1 = df1.values
    v2 = df2.values
    if len(df1) != len(df2):
        raise Exception("Cannot merge datasets of different lengths")
    for index, r1 in enumerate(v1):
        r2 = v2[index]
        if not np.array_equal(r1[:6], r2[:6]):
            raise Exception("Rows are not equal!")
        if r1[7] < 0 and r2[7] >= 0:
            df1.iat[index, 7] = r2[7]
            df1.iat[index, 8] = r2[8]
    df1.to_csv(outfile, index=False)
```
