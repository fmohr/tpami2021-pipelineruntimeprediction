```python
df = pd.read_csv("preprocessor_runtimes_dup2.csv")
```

```python
df = df.query("traintime >= 0")
df = df[df["exception"].isnull()]
```

```python
independentVariables = ['openmlid', 'algo', 'datapoints', 'trainpoints',
       'attributes', 'seed', 'numinstances_before', 'numattributes_before',
       'numlabels_before', 'numnumericattributes_before',
       'numsymbolicattributes_before', 'numberofcategories_before',
       'numericattributesafterbinarization_before',
       'attributestocover50pctvariance_before',
       'attributestocover90pctvariance_before',
       'attributestocover95pctvariance_before',
       'attributestocover99pctvariance_before', 'accvariancerel1_before',
       'accvariancerel2_before', 'accvariancerel3_before',
       'accvariancerel4_before', 'accvariancerel5_before',
       'accvariancerel6_before', 'accvariancerel7_before',
       'accvariancerel8_before', 'accvariancerel9_before',
       'accvariancerel10_before', 'accvarianceabs1_before',
       'accvarianceabs2_before', 'accvarianceabs3_before',
       'accvarianceabs4_before', 'accvarianceabs5_before',
       'accvarianceabs6_before', 'accvarianceabs7_before',
       'accvarianceabs8_before', 'accvarianceabs9_before',
       'accvarianceabs10_before', 'totalvariance_before']
targets = ['numinstances_after', 'numattributes_after', 'numlabels_after', 'numnumericattributes_after',
       'numsymbolicattributes_after', 'numberofcategories_after',
       'numericattributesafterbinarization_after',
       'attributestocover50pctvariance_after',
       'attributestocover90pctvariance_after',
       'attributestocover95pctvariance_after',
       'attributestocover99pctvariance_after', 'accvariancerel1_after',
       'accvariancerel2_after', 'accvariancerel3_after',
       'accvariancerel4_after', 'accvariancerel5_after',
       'accvariancerel6_after', 'accvariancerel7_after',
       'accvariancerel8_after', 'accvariancerel9_after',
       'accvariancerel10_after', 'accvarianceabs1_after',
       'accvarianceabs2_after', 'accvarianceabs3_after',
       'accvarianceabs4_after', 'accvarianceabs5_after',
       'accvarianceabs6_after', 'accvarianceabs7_after',
       'accvarianceabs8_after', 'accvarianceabs9_after',
       'accvarianceabs10_after', 'totalvariance_after']
```

```python
df[independentVariables]
```

```python
X = df[independentVariables]
datasets = pd.unique(df["openmlid"])
algorithms = pd.unique(df["algo"])
```

```python
def overMAPE(A, F):
    n = 0
    s = 0
    for i in range(len(A)):
        if A[i] <= F[i]:
            s += (F[i] - A[i]) / A[i]
            n+=1
    if n == 0:
        return 0
    return s / n

def underMAPE(A, F):
    n = 0
    s = 0
    for i in range(len(A)):
        if A[i] >= F[i]:
            s += (A[i] - F[i]) / F[i]
            n+=1
    if n == 0:
        return 0
    return s / n
```

```python
overestimates = {}
underestimates = {}
rsme = {}
```

```python
targets = ['numinstances_after', 'numattributes_after', 'numlabels_after', 'numnumericattributes_after',
       'numsymbolicattributes_after', 'numberofcategories_after',
       'numericattributesafterbinarization_after', 'attributestocover50pctvariance_after',
       'attributestocover90pctvariance_after', 'attributestocover95pctvariance_after',
       'attributestocover99pctvariance_after']
```

```python
from sklearn.ensemble import RandomForestRegressor
import sklearn as sk
allFeatures = ['numinstances_before', 'numattributes_before', 'numlabels_before',
       'numnumericattributes_before', 'numsymbolicattributes_before',
       'numberofcategories_before',
       'numericattributesafterbinarization_before',
       'attributestocover50pctvariance_before',
       'attributestocover90pctvariance_before',
       'attributestocover95pctvariance_before',
       'attributestocover99pctvariance_before', 'accvariancerel1_before',
       'accvariancerel2_before', 'accvariancerel3_before',
       'accvariancerel4_before', 'accvariancerel5_before',
       'accvariancerel6_before', 'accvariancerel7_before',
       'accvariancerel8_before', 'accvariancerel9_before',
       'accvariancerel10_before', 'accvarianceabs1_before',
       'accvarianceabs2_before', 'accvarianceabs3_before',
       'accvarianceabs4_before', 'accvarianceabs5_before',
       'accvarianceabs6_before', 'accvarianceabs7_before',
       'accvarianceabs8_before', 'accvarianceabs9_before',
       'accvarianceabs10_before', 'totalvariance_before']
pbar = tqdm(total = len(algorithms) * len(datasets) * len(["numattributes", "numlabels", "numnumericattributes", "numsymbolicattributes", "numberofcategories", "numericattributesafterbinarization", "attributestocover50pctvariance", "attributestocover90pctvariance", "attributestocover95pctvariance", "attributestocover99pctvariance"]))
for tStripped in ["numattributes", "numlabels", "numnumericattributes", "numsymbolicattributes", "numberofcategories", "numericattributesafterbinarization", "attributestocover50pctvariance", "attributestocover90pctvariance", "attributestocover95pctvariance", "attributestocover99pctvariance"]:
    features = [tStripped + "_before"]
    t = tStripped + "_after"
    Y = df[t]
    overestimates[t] = {}
    underestimates[t] = {}
    rsme[t] = {}
    cdf = pd.concat([X, Y], axis=1)
    for a in algorithms:
        adf = cdf.query("algo == '" + str(a) + "'")
        #print("Algo " + a)
        overestimates[t][a] = {}
        underestimates[t][a] = {}
        rsme[t][a] = {}
        for d in datasets:
            pbar.update(1)
            trainDF = adf.query("openmlid != " + str(d))
            validationDF = adf.query("openmlid == " + str(d))
            #print("\tDS" + str(d) + ". Train data size: " + str(len(trainDF)) + ". Validation data size: " + str(len(validationDF)))
            if len(validationDF) > 0:    
                reg = RandomForestRegressor(n_estimators=10)
                reg.fit(trainDF[features].values, trainDF[t].values)
                actuals = validationDF[t].values
                predictions = np.maximum(1, np.round(reg.predict(validationDF[features])))
                if len(predictions) != len(actuals):
                    raise Exception()
                #print("\t\tGT: " + str(validationDF[t].values))
                #print("\t\tBF: " + str(validationDF["numattributes_before"].values))
                #print("\t\tPR: " + str(predictions))
                error = np.sqrt(sk.metrics.mean_squared_error(actuals, predictions))
                e1 = np.round(overMAPE(actuals, predictions), 2) if np.count_nonzero(actuals >= predictions) > 0 else 0
                e2 = np.round(underMAPE(actuals, predictions), 2) if np.count_nonzero(actuals <= predictions) > 0 else 0
                overestimates[t][a][d] = e1
                underestimates[t][a][d] = e2
                rsme[t][a][d] = error
                #print("\t\t" + str(error) + "/" + str(e1) + "/" + str(e2) +" with absolute avg ground truth vals " + str(np.mean(validationDF[t].values)) + " and max " + str(np.max(validationDF[t].values)))
                #print(actuals - predictions)
            else:
                pass
                #print("\t\tno samples for ds " + str(d))
pbar.close()
```

```python
import matplotlib.pyplot as plt

def plotPerformances(rsme, overestimates, underestimates, attribute):
    fig = plt.figure()
    vals = []
    for algo in rsme[attribute]:
        d = rsme[attribute][algo]
        vals.append([d[x] for x in d])
    plt.boxplot(vals, 0, '')
    fig.axes[0].set_xticklabels(algorithms, rotation=90)
    plt.title(attribute + " - RMSE")

    fig = plt.figure()
    vals = []
    for algo in overestimates[attribute]:
        d = overestimates[attribute][algo]
        vals.append([d[x] for x in d])
    plt.boxplot(vals, 0, '')
    fig.axes[0].set_xticklabels(algorithms, rotation=90)
    plt.title(attribute + " - Overestimates")

    fig = plt.figure()
    vals = []
    for algo in underestimates[attribute]:
        d = underestimates[attribute][algo]
        vals.append([d[x] for x in d])
    plt.boxplot(vals, 0, '')
    fig.axes[0].set_xticklabels(algorithms, rotation=90)
    plt.title(attribute + " - Underestimates")

for key in rsme:
    plotPerformances(rsme, overestimates, underestimates, key)
```
